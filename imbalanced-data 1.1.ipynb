```python
def missing_values_table(df):
    missing_values = df.isin([-1]).sum()
    missing_values_percent = 100*(missing_values/len(df))
    missing_values_table = pd.concat([missing_values,missing_values_percent],axis =1)
    missing_values_table = missing_values_table.rename(columns = {0:'Missing Values',1: "Percent of Missing values"})
    return missing_values_table
```

```python
def finding_outliers(df):
    variable_name = []
    outlier_count = []
    outlier_percent = []
    df_outliers = pd.DataFrame(columns = ["Variable Name","Outlier count", "Percent of Outliers"])
    for i in range(1,df.shape[1]):
        variable = df.iloc[:,i]
        Q1 = variable.quantile(0.25)
        Q3 = variable.quantile(0.75)
        IQR = Q3- Q1
        alt_sinir = Q1 - 1.5*IQR
        ust_sinir= Q3+1.5*IQR
        outliers_count = len(df[(variable.values>ust_sinir) | (variable.values<alt_sinir)])
        variable_name.append(df.columns[i])
        outlier_count.append(outliers_count) 
        outlier_percent.append(((outliers_count/len(variable))*100))
    df_outliers["Variable Name"] = variable_name
    df_outliers["Outlier count"] = outlier_count
    df_outliers['Percent of Outliers'] = outlier_percent
    return df_outliers
```

```python
def drop_outliers(variable):
    Q1 = variable.quantile(0.25) 
    Q3 = variable.quantile(0.75)
    IQR = Q3- Q1
    alt_sinir = Q1 - 1.5*IQR
    ust_sinir= Q3+1.5*IQR
    df_new = pd.DataFrame((variable.values < ust_sinir) & (variable.values > alt_sinir))
    return df_new
```

```python
from sklearn.preprocessing import StandardScaler
pipeline_scale = Pipeline([('scaler',StandardScaler())])
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn.naive_bayes import GaussianNB as gnb
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier as dt
from sklearn.ensemble import RandomForestClassifier as rf
from sklearn.ensemble import GradientBoostingClassifier as gb
from xgboost import XGBClassifier as xgb
from catboost import CatBoostClassifier as cb
from lightgbm import LGBMClassifier as lgbm
from sklearn.model_selection import cross_val_score
```

```python
def try_model(X_train,y_train,class_weights_parameter):
   
    
    skf = StratifiedKFold(n_splits = 5, random_state = 42, shuffle = True)
    
    models = [
        
            cb(silent = True,class_weights = class_weights_parameter,task_type = 'GPU', devices = '0:1')
        
                
    ]
    
    results_df = pd.DataFrame(columns = ["model",
                                       "mean_test_score_percent",
                                        "std_test_score"
                                        ])
    for model in models:
        model_name = model.__class__.__name__
        print("Fitting:",model_name)
        
        pipeline_with_model = Pipeline([("transformer", pipeline_scale),
                                       ("model",model)])
                     
        
        scores = cross_val_score(pipeline_with_model,
                                X_train,
                                y_train,
                                cv = skf,
                                scoring  ='f1_macro')
        
        
        print(model_name," fitted Succesfully")
        
        print("Model: {} Score: {}".format(model_name, scores.mean()))
        
        results_df = results_df.append({'model':model_name,
                                       'mean_test_score_percent':(scores.mean()*100).round(3),
                                       'std_test_score':scores.std()},
                                        ignore_index = True)
    return results_df
```

```python
try_model(X_train,y_train,None)
```

SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
def SMOTE_process(X_train,y_train):
    sm = SMOTE()
    X_res, y_res = sm.fit_resample(X_train,y_train)
    return X_res,y_res
```

```python
X_res,y_res = SMOTE_process(X_train,y_train)
try_model(X_res,y_res,None)
```

Under Sampling

```python
def Under_sampling(X_train,y_train):
    from imblearn.under_sampling import RandomUnderSampler
    rus = RandomUnderSampler(random_state=0)
    X_resampled, y_resampled = rus.fit_resample(X_train, y_train)
    return X_resampled,y_resampled
```

```python
X_res,y_res = Under_sampling(X_train,y_train)
try_model(X_res,y_res,None)
```
